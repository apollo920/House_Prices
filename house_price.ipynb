```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)

df = pd.read_csv('train.csv')
df.head()
```

```python
df.info()
missing = df.isnull().sum()
missing[missing > 0].sort_values(ascending=False)
```

- Analisar a relação entre algumas viriáveis e o preço(SalePrice). As 3 mais importantes inicialmente são GrLlivArea, OverrallQual, YearBuilt.

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[['SalePrice', 'GrLivArea', 'OverallQual', 'YearBuilt']].corr(), annot=True, cmap='viridis')
plt.title('Correlação com o preço de venda')
plt.show()

sns.scatterplot(x='GrLivArea', y='SalePrice', data=df)
plt.title('Área Útil vs Preço de Venda')
plt.show()

sns.boxplot(x='OverallQual', y='SalePrice', data=df)
plt.title('Qualidade da construção vs Preço de Venda')
plt.show()
```

- Tratamento de Outliers

```python
outliers = df[(df['GrLivArea'] > 4000) & (df['SalePrice'] < 300000)]
print('Outliers detectados')
display(outliers[['Id','GrLivArea','SalePrice']])

df=df.drop(outliers.index)

print(f'Novo tamanho do dataset: {df.shape}')
```
